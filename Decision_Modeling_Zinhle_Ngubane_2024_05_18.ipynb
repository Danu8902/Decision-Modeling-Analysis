```python
import pandas as pd
import numpy as np

def time_to_seconds(time_str):
  if pd.isna(time_str):
    return np.nan
    mins, secs = time_str.split(':')
    return int(mins) * 60 + float(secs)

driver_standings_df = pd.read_csv('driver_standings.csv')
drivers_df = pd.read_csv('drivers.csv')
lap_times_df = pd.read_csv('lap_times.csv')
qualifying_df = pd.read_csv('qualifying.csv')
results_df = pd.read_csv('results.csv')
status_df = pd.read_csv('status.csv')

print("Initial Driver Standings DataFrame:")
print(driver_standings_df.head())

print("Initial Drivers DataFrame:")
print(drivers_df.head())

print("Initial Lap Times DataFrame:")
print(lap_times_df.head())

print("Initial Qualifying DataFrame:")
print(qualifying_df.head())

print("Initial Results DataFrame:")
print(results_df.head())

print("Initial Status DataFrame:")
print(status_df.head())

# Cleaning Driver Standings DataFrame
print("Missing values in Driver Standings DataFrame:")
print(driver_standings_df.isnull().sum())

# Assuming we need to convert necessary columns to appropriate data types
driver_standings_df['points'] = pd.to_numeric(driver_standings_df['points'], errors = 'coerce')

# Cleaning Drivers DataFrame
print("Missing values in Drivers DataFrame:")
print(drivers_df.isnull().sum())

# Cleaning Lap Times DataFrame
lap_times_df = lap_times_df.apply(pd.to_numeric, errors = 'ignore')

# Cleaning Qualifying DataFrame
qualifying_df['q1'] = qualifying_df['q1'].apply(time_to_seconds)
qualifying_df['q2'] = qualifying_df['q2'].apply(time_to_seconds)
qualifying_df['q3'] = qualifying_df['q3'].apply(time_to_seconds)

# Cleaning Results DataFrame
results_df = results_df.apply(pd.to_numeric, errors = 'ignore')
results_df['position_x'] = pd.to_numeric(results_df['position'], errors = 'coerce')

# Cleaning Status DataFrame
status_df.drop_duplicates(inplace = True)
status_df.dropna(inplace = True)

print("Cleaned Driver Standings DataFrame:")
print(driver_standings_df.head())

print("Cleaned Drivers DataFrame:")
print(drivers_df.head())

print("Cleaned Lap Times DataFrame:")
print(lap_times_df.head())

print("Cleaned Qualifying DataFrame:")
print(qualifying_df.head())

print("Cleaned Results DataFrame:")
print(results_df.head())

print("Cleaned Status DataFrame:")
print(status_df.head())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, accuracy_score, confusion_matrix, classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

driver_standings_df = pd.read_csv('driver_standings.csv')
drivers_df = pd.read_csv('drivers.csv')
lap_times_df = pd.read_csv('lap_times.csv')
qualifying_df = pd.read_csv('qualifying.csv')
results_df = pd.read_csv('results.csv')
status_df = pd.read_csv('status.csv')

dfs = [driver_standings_df, drivers_df, lap_times_df, qualifying_df, results_df, status_df]
for df in dfs:
  print(df.columns)

# Combine all relevant datasets into a single DataFrame
data = results_df.merge(driver_standings_df, on = ['raceId', 'driverId'], suffixes = ('_result', '_standings'))
data = data.merge(lap_times_df, on = ['raceId', 'driverId'])
data = data.merge(qualifying_df, on = ['raceId', 'driverId'])

# Drop any duplicates or irrelevant columns
data.drop_duplicates(inplace = True)

# Inspect the merged data
print(data.head())

# Function to compute statistics only on numeric columns
def compute_statistics(df):
    numeric_df = df.select_dtypes(include = [np.number])
    if numeric_df.empty:
      return None, None, None, None
    mean = numeric_df.mean()
    median = numeric_df.median()
    std_dev = numeric_df.std()
    summary_stats = numeric_df.describe()
    return mean, median, std_dev, summary_stats

# Compute Summary Statistics for merged data
mean, median, std_dev, summary_stats = compute_statistics(data)
print("Means:\n", mean)
print("\nMedians:\n", median)
print("\nStatistics:\n", std_dev)
print("\n Summary Statistics:\n", summary_stats)

# Plot distributions of relevant features
plt.figure(figsize = (15, 10))
sns.histplot(data['positionOrder'], kde =True)
plt.title('Distributions of Final Positions')
plt.xlabel('Final Position')
plt.ylabel('Frequency')
plt.show()

# Scatter plot of qualifying time vs. final position
plt.figure(figsize = (15, 10))
sns.scatterplot(data = data, x = 'milliseconds_y', y = 'positionOrder')
plt.title('Qualifying Time vs Final Position')
plt.xlabel('Qualifying Time')
plt.ylabel('Final Position')
plt.show()

# Performance Hypothesis
# Select relevant features for regression
x = data[['milliseconds_y']]
y = data['positionOrder']

# Split the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

# Initialize and train the data
model = LinearRegression()
model.fit(x_train, y_train)

# Make predictions
y_pred = model.predict(x_test)

# Evaluate the model
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'MAE: {mae}')
print(f'MSE: {mse}')
print(f'R2 Score: {r2}')

# Plot the results
plt.scatter(y_test, y_pred)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Linear Regression Predictions vs Actual')
plt.show()

# Experience Hypothesis
# Select relevant features and target
x = data[['milliseconds_y']]
y = data['positionOrder']

# Split the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

# Initialize and train the model
model = DecisionTreeClassifier()
model.fit(x_train, y_train)

# Make predictions
y_pred = model.predict(x_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print(f'Confusion Matrix: \n{cm}')
print(f'Classification Report: \n{classification_report(y_test, y_pred)}')

# Plot the confusion matrix
plt.figure(figsize = (10, 7))
sns.heatmap(cm, annot = True, fmt = 'd')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Consistency Hypothesis
# Select relevant features and target
x = data[['milliseconds_y']]
y = data['positionOrder']

# Split the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

# Initialize and train the model
model = RandomForestClassifier()
model.fit(x_train, y_train)

# Make predictions
y_pred = model.predict(x_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print(f'Confusion Matrix: \n{cm}')
print(f'Classfication Report: \n{classification_report(y_test, y_pred)}')

# Plot the confusion matrix
plt.figure(figsize = (10, 7))
sns.heatmap(cm, annot = True, fmt = 'd')
plt.title('Confusion Matrix 2')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Export the data
export_data = data[['raceId', 'driverId', 'milliseconds_y', 'positionOrder']].copy()
export_data.rename(columns = {
    'milliseconds_y' : 'QualifyingTime',
    'positionOrder' : 'FinalPosition'
}, inplace = True)

export_data.to_csv('race_analysis.csv', index = False)
```
